```python
RPMS = ["DIAPH1", "ABCA2", "CXCR4", "AKAP1", "LAMC2", "HIC2", "BAG2", "FADS3", "PIM1", "TRAPPC3", "PSMD4", "ARID3B", "PAPPA", "PPP1R15A", "RDX", "DMD", "BACH1", "DYRK1A", "GPR56", "PEBP1", "EIF4G1", "MARK1", "LAMB3", "BRD2", "MAP4K4", "OR2F1", "S100A10", "UBE2E3", "TNFRSF12A", "DYSF", "PHLDA2", "HMGA2", "IRS2", "HCN3", "KRT16", "ENO2", "IGF2BP3", "IGF2BP2", "FOXN1", "RGS2", "RPL23A", "ELK3", "CAPN6", "GOLT1B", "GKN1", "RAB3D", "GNPTAB", "PCDH17", "KCNA2", "GTPBP2", "BMP2", "COL1A2", "HOXB3", "LMNA", "NDRG2", "COL3A1", "SNCB", "ADRA1A", "OSBP", "CDV3", "BTK", "GFAP", "IL13", "DSC2", "COL7A1", "FAP", "ELAVL2", "IL1RN", "MMP1", "SHC3", "EN1", "ATP1B1", "HMGA1", "DLG4", "PKP3", "TGFBR1", "PSMD7", "KRT8", "ENO1", "LPP", "COL5A2", "CA7", "MMP3", "APOBEC1", "SQSTM1", "F2RL2", "GIT1", "RAB30", "TIAL1", "GPR3", "EPHB2", "LRIG2", "GNAI1", "ABCD1", "EEF1A1", "NRAS", "MPV17", "SPP1", "DCTN2", "NR1I3", "LAMC1", "BDKRB1", "UBE2C", "AKT3", "TAGLN2", "SCRN1", "EMP3", "SYNGR1", "PADI4", "LTB4R2", "PSME4", "MAST2", "RPS6KA4", "NF2", "MAP2", "CDKN1A"]
BPMS = ["IL7", "DYM", "HS3ST3B1", "BMPER", "MAGEC1", "PDE1C", "FBXO42", "PRDM1", "FRMPD4", "IL1RAP", "RCAN3", "MYCT1", "HERC3"]
```

```python
from gseapy import ssgsea

# Define gene sets
genesets = {'RPMS': RPMS, 'BPMS': BPMS}

# Calculate ssGSEA scores for both RPMS and BPMS
ssgsea_score_table = ssgsea(data=data, gene_sets=genesets, outdir=None, min_size = 5)
```

```python
ssgsea_score_table.res2d
```

```python
# Pivot the DataFrame to create separate columns for RPMS and BPMS
ssgsea_score_table = ssgsea_score_table.res2d.pivot(
    index='Name',
    columns='Term',
    values='NES'
).reset_index()

# Rename the columns
ssgsea_score_table.columns.name = None 
ssgsea_score_table = ssgsea_score_table.rename(columns={
    'RPMS': 'RPMS_NES',
    'BPMS': 'BPMS_NES'
})

ssgsea_score_table.head()
```

```python
ssgsea_score_table
```

```python
# Calculate the median values for RPMS_NES and BPMS_NES
median_rpms = ssgsea_score_table['RPMS_NES'].median()
median_bpms = ssgsea_score_table['BPMS_NES'].median()

# Classify each row based on the median values
ssgsea_score_table['Type'] = ''
ssgsea_score_table.loc[(ssgsea_score_table['RPMS_NES'] >= median_rpms) & (ssgsea_score_table['BPMS_NES'] >= median_bpms), 'Type'] = 'RPMS+/BPMS+'
ssgsea_score_table.loc[(ssgsea_score_table['RPMS_NES'] < median_rpms) & (ssgsea_score_table['BPMS_NES'] >= median_bpms), 'Type'] = 'RPMS-/BPMS+'
ssgsea_score_table.loc[(ssgsea_score_table['RPMS_NES'] >= median_rpms) & (ssgsea_score_table['BPMS_NES'] < median_bpms), 'Type'] = 'RPMS+/BPMS-'
ssgsea_score_table.loc[(ssgsea_score_table['RPMS_NES'] < median_rpms) & (ssgsea_score_table['BPMS_NES'] < median_bpms), 'Type'] = 'RPMS-/BPMS-'

ssgsea_score_table.head()
```

```python
survival_data = pd.read_csv("./Transcriptomics/brca_metabric/data_clinical_patient.txt", sep='\t', header = 4)

survival_data
```

```python
survival_data_subset = survival_data[["PATIENT_ID", "OS_MONTHS", "OS_STATUS", "RFS_MONTHS", "RFS_STATUS"]]
survival_data_subset = survival_data_subset.dropna()
survival_data_subset
```

```python
final_data = pd.merge(ssgsea_score_table, survival_data_subset, left_on='Name', right_on='PATIENT_ID')

final_data = final_data.drop(["PATIENT_ID"], axis=1)

# Convert OS_STATUS to binary (0 or 1)
final_data['OS_STATUS'] = final_data['OS_STATUS'].str[0].astype(int)

# Convert RFS_STATUS to binary (0 or 1)
final_data['RFS_STATUS'] = final_data['RFS_STATUS'].str[0].astype(int)

final_data
```

```python
from lifelines import KaplanMeierFitter
from lifelines.statistics import multivariate_logrank_test

import matplotlib.pyplot as plt

# Set up the KaplanMeierFitter
kmf = KaplanMeierFitter()

# Set up the figure
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)

# Plot for OS
for group in final_data['Type'].unique():
    mask = final_data['Type'] == group
    kmf.fit(
        final_data[mask]['OS_MONTHS'],
        final_data[mask]['OS_STATUS'],
        label=group
    )
    kmf.plot(ci_show=False)

plt.title('BRCA METABRIC Overall Survival')
plt.xlabel('Months')
plt.ylabel('Survival probability')
plt.grid(True)

# Plot for RFS
plt.subplot(1, 2, 2)

for group in final_data['Type'].unique():
    mask = final_data['Type'] == group
    kmf.fit(
        final_data[mask]['RFS_MONTHS'],
        final_data[mask]['RFS_STATUS'],
        label=group
    )
    kmf.plot(ci_show=False)

plt.title('BRCA METABRIC Relapse Free Status')
plt.xlabel('Months')
plt.ylabel('Survival probability')
plt.grid(True)

plt.tight_layout()
```

```python
import pandas as pd
import numpy as np
from lifelines import CoxPHFitter
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare the data
y = final_data[["Name", "OS_STATUS", "OS_MONTHS", "Type"]].copy().dropna(subset=['OS_MONTHS', 'Type'])

# Set reference category 
y["Type"] = pd.Categorical(y["Type"], categories=["RPMS-/BPMS-", "RPMS+/BPMS+", "RPMS+/BPMS-", "RPMS-/BPMS+"])

# Fit Cox model
cph = CoxPHFitter()
cph.fit(y, duration_col="OS_MONTHS", event_col="OS_STATUS", formula="Type")

# Get summary statistics
summary = cph.print_summary()

# Create DataFrame with hazard ratios and confidence intervals
results_df = pd.DataFrame({
    'HR': np.exp(cph.params_),
    'HR.confint.lower': np.exp(cph.confidence_intervals_.iloc[:,0]),
    'HR.confint.upper': np.exp(cph.confidence_intervals_.iloc[:,1]),
    'p.value': cph.summary['p']
})

# Round values to 2 decimal places
results_df = results_df.round(2)

# Add labels column
results_df['labels'] = results_df.index

# Create the forest plot
plt.figure(figsize=(10, 6))
sns.set_style("white")

# Plot points and error bars
for idx, row in results_df.iterrows():
    color = 'red' if row['HR'] > 1 else 'blue'
    plt.errorbar(x=row['HR'], y=row['labels'],
                xerr=[[row['HR'] - row['HR.confint.lower']], 
                      [row['HR.confint.upper'] - row['HR']]],
                fmt='o', color=color, capsize=5, capthick=2, 
                markersize=8, elinewidth=2)

# Add significance stars
for idx, row in results_df.iterrows():
    if row['p.value'] < 0.001:
        star = '***'
    elif row['p.value'] < 0.01:
        star = '**'
    elif row['p.value'] < 0.05:
        star = '*'
    else:
        star = ''
    
    y_pos = row['labels']
    x_pos = row['HR.confint.upper'] + 0.1
    plt.text(x_pos, y_pos, star, fontsize=12, va='center')

# Add reference line
plt.axvline(x=1, color='black', linestyle='--', alpha=0.5)

# Customize plot
plt.xlabel('Hazard Ratio', fontsize=12, fontweight='bold')
plt.ylabel('')
plt.title(f'For BRCA METABRIC Overall Survival \nLog-Rank p: {cph.log_likelihood_ratio_test().p_value:.2f}', 
          fontsize=10, fontweight='bold')

# Adjust layout
plt.tight_layout()
plt.show()
```
